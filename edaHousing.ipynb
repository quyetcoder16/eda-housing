```python
# FUNCTION FOR BI - VARIENCE WITH PRICE :

def bi_varient_price(x):
  sns.barplot(data = df , x = x, y = 'price' , color = 'lightblue' , edgecolor = 'black')
  plt.xlabel(x.upper())
  plt.ylabel('PRICE IN MILLIONS')
  plt.title('REALTION OF PRICE AND '+str(x.upper()))
```

```python
bi_varient_price('bedrooms')
```

```python
bi_varient_price("stories")
```

```python
# Check Column Area
Q1 = df['area'].quantile(0.25)
Q3 = df['area'].quantile(0.75)

IQR = Q3 - Q1

# identify outliers
threshold = 1.5
df[(df['area'] < Q1 - threshold * IQR) | (df['area'] > Q3 + threshold * IQR)]
```

```python
# delete the outlier datas

df.drop(df[(df['area'] < Q1 - threshold * IQR) | (df['area'] > Q3 + threshold * IQR)].index, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

encoded_classes = []
encoder = LabelEncoder()

column = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea', 'furnishingstatus']

for col in column:
    df[col] = encoder.fit_transform(df[col])
    encoded_classes.append(encoder.classes_)
```
